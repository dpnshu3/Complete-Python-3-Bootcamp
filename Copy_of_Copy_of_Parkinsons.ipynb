```python
from scipy.io import arff
import pandas as pd 
import numpy as np
dataset = arff.loadarff('./Dataset/php4ylQmK.arff')[0]
df = pd.DataFrame(dataset)
df
```

## As you can see above the values in `Class` are string so we need to convert them to integer

```python
cl = []
for c in df['Class']:
  cl.append(int(c))
df['Class'] = cl
```

```python
df
```

#Separate the features and class

```python
    features = df.loc[:, 'V1':'V22']
    classes = df['Class']
    X = features.values
    Y = classes.values
```

#Import kfold from sklearn
https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html#sklearn.model_selection.KFold 

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10,shuffle=True)
```

#We need multivariate normal random variable
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multivariate_normal.html


```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score
from scipy.stats import multivariate_normal

accuracy_gda = []
balanced_accuracy_gda = []

accuracy_lda = []
balanced_accuracy_lda = []

accuracy_qda = []
balanced_accuracy_qda = []

for train, test in kfold.split(X):
      #Split the dataset for traning and testing
      X_train, X_test = X[train], X[test]
      Y_train, Y_test = Y[train], Y[test]

      healthy = []
      unhealthy = []
      for k, v  in enumerate(Y_train):
        if v == 1:
          healthy.append(k)
      for k, v in enumerate(Y_train):
        if v == 2:
          unhealthy.append(k)

      prior_healthy = len(healthy)/len(Y_train)
      prior_unhealthy = 1 - prior_healthy 

      #constuct new data sent of healthy and unhealthy
      new_healthy = X_train[healthy, :]
      new_unhealthy = X_train[unhealthy, :]

      if len(new_healthy) > 0:
        mean_healthy = np.mean(new_healthy, axis=0)
      if len(new_unhealthy) > 0:
        mean_unhealthy = np.mean(new_unhealthy, axis=0)

      #covarance matrix
      sigma = np.cov(X_train, rowvar=False) # for gda
      inv_sigma = np.linalg.pinv(sigma) #for lda
      # for qda sigma calculation
      sigma_healthy = np.cov(new_healthy, rowvar=False)
      sigma_unhealthy = np.cov(new_unhealthy, rowvar=False)

      gda_predicted = []
      lda_predicted = []
      qda_predicted = []

      #GDA Calculation
      for x in X_test:
            multivariate_healthy =  multivariate_normal(mean_healthy, sigma, allow_singular=True)
            multivariate_unhealthy =  multivariate_normal(mean_unhealthy, sigma, allow_singular=True)
            PDF_healthy = multivariate_healthy.pdf(x)
            PDF_unhealthy = multivariate_unhealthy.pdf(x)
            
            if ((abs(PDF_healthy - PDF_unhealthy) < 0.001) and (PDF_healthy > 0.01)):
                gda_predicted.append(1)
            else:
                gda_predicted.append(2)

      accuracy_gda.append(accuracy_score(Y_test, gda_predicted))
      balanced_accuracy_gda.append(balanced_accuracy_score(Y_test, gda_predicted))

      #LDA Calculation          
      for x in X_test:
            #calculate dlt for both the classes
            dlt_healthy = (np.dot(np.dot(x.T, inv_sigma), mean_healthy)) - (0.5 * np.dot(np.dot(mean_healthy.T, inv_sigma), mean_healthy)) + np.log(prior_healthy)
            dlt_unhealthy = (np.dot(np.dot(x.T, inv_sigma), mean_unhealthy)) - (0.5 * np.dot(np.dot(mean_unhealthy.T, inv_sigma), mean_unhealthy)) + np.log(prior_unhealthy)

            if dlt_healthy - dlt_unhealthy > 0:
                lda_predicted.append(1)
            else:
                lda_predicted.append(2)
      accuracy_lda.append(accuracy_score(Y_test, lda_predicted))
      balanced_accuracy_lda.append(balanced_accuracy_score(Y_test, lda_predicted))

      # QDA Calculation
      for X_new in X_test:
            #calculate dlt for both the classes
            dlt_healthy = -0.5 * np.log(np.linalg.det(sigma_healthy)) - 0.5 * np.dot(np.dot((X_new - mean_healthy).T, np.linalg.pinv(sigma_healthy)), (X_new - mean_healthy)) + np.log(prior_healthy)
            dlt_unhealthy = -0.5 * np.log(np.linalg.det(sigma_unhealthy)) - 0.5 * np.dot(np.dot((X_new - mean_unhealthy).T, np.linalg.pinv(sigma_unhealthy)), (X_new - mean_unhealthy)) + np.log(prior_unhealthy)
 
            if dlt_healthy - dlt_unhealthy > 0:
                qda_predicted.append(1)
            else:
                qda_predicted.append(2)
      accuracy_qda.append(accuracy_score(Y_test, qda_predicted))
      balanced_accuracy_qda.append(balanced_accuracy_score(Y_test, qda_predicted))
```

```python
acc_df = pd.DataFrame()
acc_df['Cross-Validation'] = ['CV1','CV2','CV3','CV4','CV5','CV6','CV7','CV8','CV9','CV10', 'Mean']
accuracy_gda.append(np.mean(accuracy_gda))
acc_df['GDA']= accuracy_gda
accuracy_lda.append(np.mean(accuracy_lda))
acc_df['LDA'] = accuracy_lda
accuracy_qda.append(np.mean(accuracy_qda))
acc_df['QDA'] = accuracy_qda
acc_df
```

```python
bal_acc_df = pd.DataFrame()
bal_acc_df['CROSS-Validation'] = ['CV1','CV2','CV3','CV4','CV5','CV6','CV7','CV8','CV9','CV10', 'Mean']
balanced_accuracy_gda.append(np.mean(balanced_accuracy_gda))
bal_acc_df['GDA'] = balanced_accuracy_gda
balanced_accuracy_lda.append(np.mean(balanced_accuracy_lda))
bal_acc_df['LDA'] = balanced_accuracy_lda
balanced_accuracy_qda.append(np.mean(balanced_accuracy_qda))
bal_acc_df['QDA'] = balanced_accuracy_qda
bal_acc_df

```
